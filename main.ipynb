```python
import pandas as pd
import pandas_profiling
import numpy as np
```

```python
# Displaying dataset with all columns
pd.set_option('display.max_columns', None)
dataset = pd.read_csv('Botswana_Data.csv')
dataset
```

```python
# Dropping 'Date' column as it has no use in the model
df = dataset.drop('Date', axis=1)
df
```

```python
# Making profile report for better understanding of the dataset
profile = pandas_profiling.ProfileReport(df)
# profile.to_file("Botswana_Data_report.html")
```

```python
col_list = df.columns.values
col_list
```

```python
# Removing columns with only one type of value (same value) thoughout the row, again no use.
unique_value_dict = []
count_one_list = []
for c in col_list:
    unique_value_count = len(df[c].unique())
    unique_value_dict.append(f'{c} = {unique_value_count}') 
    if unique_value_count == 1:
        df.drop(c, axis=1, inplace = True)
        count_one_list.append(c)
```

```python
# Printing columns with numbers of unique values (different values) it has.
print(unique_value_dict,'\n', count_one_list)        
df # dataset after dropping columns with only one value (same value) through out the row
```

```python
df.corr() # Displaying the correlation of columns having numeric value with eachother to see the effect of each feature
```

```python
# Removing of highly correlated columns as they will result in overfitting decause they have almost same value corresponding to its corelated column/s 

df = df.drop('Subheading', axis=1)
df = df.drop('Heading', axis=1)
df = df.drop('Gross Weight', axis=1)
df = df.drop('Customs value USD', axis=1)
df = df.drop('Vat', axis=1)

df.corr()
```

```python
df.info()
```

```python
# categorical (don't have numeric value) columns list
cat_col = list(df.loc[:,df.dtypes==np.object].columns)
cat_col
```

```python
# Changing categorical values to integer as model can only handel numerical data
from sklearn.preprocessing import LabelEncoder

data_dict = {}

for col in cat_col:
    data_dict['le_'+col] = LabelEncoder()
    df[col] = data_dict['le_'+col].fit_transform(df[col])
    print(col,df[col].unique())
```

```python
# printing the top features which will contribute the most (to Freight BWP) for accuracy of the model
from sklearn.ensemble import ExtraTreesRegressor

labels = df.columns.drop(['Freight BWP'])

TOP_FEATURES = 10

forest = ExtraTreesRegressor(n_estimators=5, max_depth=5, random_state=1)
forest.fit(df[labels], df['Freight BWP'])

importances = forest.feature_importances_
std = np.std([tree.feature_importances_ for tree in forest.estimators_], axis=0)
indices = np.argsort(importances)[::-1]
indices = indices[:TOP_FEATURES]

print('Top features:')
for f in range(TOP_FEATURES):
    print(f'{f+1}. feature {indices[f]} ({importances[indices[f]]:.5f})')
```

```python
X = df.drop("Freight BWP", axis = 1) # Input data
Y = df["Freight BWP"] # Output data (predicting data)

display(X, Y)
```

```python
# fitting and prediting
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

model = LinearRegression() # Best for small datasets 
reg = model.fit(X,Y.values)
Y_pred = model.predict(X)
print('Predicted value:', list(Y_pred),'\n\n' ,f"Original value:\n{df['Freight BWP']}") 
acc = reg.score(X, Y_pred) 
print(f'Accuracy: {acc*100:.0f}% Over fitted')
# I know other predictions were also asked but they will result the same (wrong predictions)
```
